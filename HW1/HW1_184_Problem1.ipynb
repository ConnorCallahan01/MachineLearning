## 1. Load the dataset from ‘train.csv’ into a pandas DataFrame and print it along with its shape.

```python
import pandas as pd
import csv

train_data = pd.read_csv('train.csv') #Loading dataset
train_data_df = pd.DataFrame(train_data) #Loading csv into dataframe
train_data_df #printing the DataFrame
```

## 2. For dealing with missing values, delete a column with any Null value (Note: not recommended approach, but here focus is one-hot encoding). Also delete the ‘ID’ column.

```python
# Print the data frame
del train_data_df['Id'] #deleting the Id column
train_data_df
```

## 3. For simplicity, we will work with all the numerical columns but just a subset of the categorical columns. Among all the remaining categorical columns after step 2, choose the categorical columns that have less than 10 unique values.

```python
# Go through the dataframe. If the type of the column is categorical, count the amount of unique values. 
# If the value of unique values is greater than 10, delete the column from the dataframe.  
for column in train_data_df: 
    if train_data_df[column].dtype != int: #if the data is not an integer value (meaning it is categorical)
        unique = len(pd.unique(train_data_df[column])) #find the count of unique values in the column
        if unique > 10: #if the count of unique is less than 10
            del train_data_df[column] #delete the column 
```

## 5. Divide the data into X and Y, where X is the set of features and Y is the target variable. For this dataset the ‘SalePrice’ is the target variable.

## 6. For step 5, two approaches will be used (as discussed in class):
- a. Set of ONLY numerical features and dropping all Categorical features
- b. Set of features comprising both numerical and one-hot encoded categorical features.

```python
from sklearn.preprocessing import OneHotEncoder
# first, initialize the target value: 
target = train_data_df['SalePrice']
del train_data_df['SalePrice']
# initialize only numerical features dataframe:
num_only_df = pd.DataFrame()
for column in train_data_df:
    if train_data_df[column].dtype == int: #if the column value is numerical
        num_only_df[column] = train_data_df[column] 

num_only_df
```

```python
import numpy as np
# now do one hot encoding on categorical data
final_one_hot = pd.DataFrame() #initialize empty data set
for column in train_data_df:
    if train_data_df[column].dtype != int: #if the column value is categorical
        column_type = train_data_df[column] #take the name of the column
        column_df = pd.DataFrame(column_type, columns=[column]) #create a new dataframe for the one hot encoding of categorical data
        enc_df = pd.get_dummies(column_df, prefix=[column+'_']) # make a new dataframe with one hot encoding of the categorical variable with new column names
        column_df = column_df.join(enc_df) # join the two new dataframes
        del column_df[column] #delete the categorical column
        final_one_hot = pd.concat([final_one_hot, column_df], axis=1) #add the one hot encoding values to the final data frame

final_one_hot #print the data frame to show it worked. 

```

## 7. Print the original dataframe joined with the one-hot encoded columns.

```python
num_onehot_df = pd.DataFrame() # initialize a new final dataframe with one hot encoded columns joined with numerical
num_onehot_df = pd.concat([num_only_df, final_one_hot], axis=1) #join them
num_onehot_df #print the combined
```

## 8. Once, the two sets of features are obtained from step 6, train a simple Linear Regression model from sklearn and obtain the Mean Absolute Error for both the cases:
- a. Mean Absolute Error when Dropping Categoricals:
- b. Mean Absolute Error with One-Hot Encoding

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

X = num_only_df
Y = target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y)

model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
print("Mean Absolute Error when Dropping Categoricals: ",metrics.mean_absolute_error(Y_test, Y_pred))
```

```python
X = num_onehot_df
Y = target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y)

model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
print("Mean Absolute Error with One-Hot Encoding: ",metrics.mean_absolute_error(Y_test, Y_pred))
```
